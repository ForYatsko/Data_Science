```python
import numpy as np
import os
import logging
from logging_config import setup_logging
from data_processing import load_data, describe_data, preprocess_data, ensure_directories, eda
from visualization import plot_predictions, visualize_statistics, plot_anomalies_with_model, visualize_metrics
from model_training import train_models, train_arima, train_lstm, create_ensemble
from anomaly_detection import detect_anomalies
from improvements import preprocess_with_anomalies, weighted_ensemble
from sklearn.model_selection import train_test_split
import matplotlib
%matplotlib inline

def main():
    # Настройка логирования
    logger = setup_logging(log_file="my_log.log", log_level=logging.DEBUG)
    logger.info("Логирование настроено.")
    ensure_directories()

    # Этап 1: Загрузка данных
    try:
        logger.info("Загрузка данных...")
        df = load_data("data/coin_Bitcoin.csv")
        logger.info("Данные успешно загружены.")
    except Exception as e:
        logger.error(f"Ошибка при загрузке данных: {e}")
        return

    # Этап 2: Предварительная обработка данных
    try:
        logger.info("Выполнение предварительной обработки данных...")
        df = preprocess_data(df)  # Обработка данных
        logger.info("Предварительная обработка завершена.")
    except Exception as e:
        logger.error(f"Ошибка при обработке данных: {e}")
        return

    # Этап 3: Анализ обработанных данных (EDA)
    try:
        logger.info("Выполнение анализа обработанных данных (EDA)...")
        eda(df)  # Анализ данных после обработки
        logger.info("EDA завершен.")
    except Exception as e:
        logger.error(f"Ошибка при выполнении EDA: {e}")
        return

    # Этап 4: Описательная статистика
    try:
        logger.info("Выполнение описательной статистики...")
        describe_data(df)
        visualize_statistics(df)
        logger.info("Описательная статистика завершена.")
    except Exception as e:
        logger.error(f"Ошибка при выполнении описательной статистики: {e}")
        return

    # Этап 5: Разделение данных на обучающую и тестовую выборки
    try:
        logger.info("Разделение данных на обучающую и тестовую выборки...")
        X = df[["Close_shift_1", "Close_shift_2", "Close_shift_3"]]
        y = df["Close"]
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
        logger.info("Разделение данных завершено.")
    except Exception as e:
        logger.error(f"Ошибка при разделении данных: {e}")
        return

    # Этап 6: Обучение моделей и создание ансамбля
    try:
        logger.info("Обучение моделей...")

        # Обучение XGBoost, LightGBM, CatBoost
        results_ml, metrics_ml = train_models(X_train, y_train, X_test, y_test)

        # Обучение ARIMA
        results_arima, metrics_arima = train_arima(y_train, y_test)

        # Обучение LSTM
        results_lstm, metrics_lstm = train_lstm(X_train, y_train, X_test, y_test)

        # Создание ансамбля
        ensemble_predictions, metrics_ensemble = create_ensemble(
            [results_ml, results_arima, results_lstm], y_test
        )
        logger.info("Ансамбль моделей создан.")

        # Объединение всех метрик
        all_metrics = {
            **metrics_ml,
            "ARIMA": metrics_arima,
            "LSTM": metrics_lstm,
            "Ensemble": metrics_ensemble
        }
    except Exception as e:
        logger.error(f"Ошибка во время обучения моделей: {e}")
        return

    # Этап 7: Визуализация прогнозов
    try:
        logger.info("Визуализация прогнозов...")

        # Проверка размеров y_test и ensemble_predictions
        if len(y_test) != len(ensemble_predictions):
            logger.error(f"Длины y_test ({len(y_test)}) и ensemble_predictions ({len(ensemble_predictions)}) не совпадают!")
            return

        # Фильтрация метрик для визуализации
        filtered_metrics = {
            key: value['MSE'] for key, value in all_metrics.items() if isinstance(value, dict) and 'MSE' in value
        }

        # Вызов функции визуализации
        plot_predictions(
            y_test,
            ensemble_predictions,
            "Прогноз ансамбля моделей",
            "ensemble_predictions.png",
            metrics=filtered_metrics  # Передача только отфильтрованных метрик
        )
        logger.info("Визуализация прогнозов завершена.")
    except Exception as e:
        logger.error(f"Ошибка при визуализации прогнозов: {e}")
        return

        # Этап 8: Поиск аномалий
        logger.info("Поиск аномалий...")
        anomalies = detect_anomalies(y_test, ensemble_predictions)
        logger.info(f"Аномалии:\n{anomalies}")
    except Exception as e:
        logger.error(f"Ошибка во время обучения моделей или анализа: {e}")

    # Этап 8: Поиск и визуализация аномалий
    try:
        logging.info("Поиск аномалий с использованием модуля аномалий...")

        # Установка порога для метода detect_anomalies
        threshold = 1000  # Порог отклонения
        logging.info(f"Используемый порог для аномалий: {threshold}")

        # Поиск аномалий с использованием функции detect_anomalies
        anomalies_series = detect_anomalies(y_test, ensemble_predictions, threshold=threshold)
        anomalies_bool = y_test.index.isin(anomalies_series.index)  # Преобразуем в логический массив
        num_anomalies = len(anomalies_series)  # Количество аномалий

        # Логирование количества аномалий
        logging.info(f"Найдено {num_anomalies} аномалий из {len(y_test)} наблюдений.")

        # Построение графика с аномалиями
        plot_anomalies_with_model(
            y_true=y_test,
            y_pred=ensemble_predictions,
            anomalies=anomalies_bool,
            title="График с аномалиями и моделью",
            filename="anomalies_with_model.png"
        )
        logging.info("Визуализация аномалий завершена.")
    except Exception as e:
        logging.error(f"Ошибка при поиске или визуализации аномалий: {e}")
        raise

if __name__ == "__main__":
    main()
```
